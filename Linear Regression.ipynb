#### R^2 = 1 - (RSS/TSS) 

#### Adjusted R^2 = (1-((1-R^2)*(N-1)/(N-P-1)))

#### N = Number of Rows
#### P = Number of predictors/columns

#### VIF (Variance Inflation Factor) = 1/(1-R^2)

#### If VIF > 10 ---> highly multi-collinear

#### Controlling error term to generalize the model is called as regularization.

#### LASSO (L1 regularization) = RSS + (lambda*sum(m))

#### Ridge Regression (L2) = RSS + (lambda*sum(m^2))

#### Elastic Net 

```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge, Lasso, RidgeCV, LassoCV, ElasticNet, ElasticNetCV, LinearRegression
from sklearn.model_selection import train_test_split
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df1 = pd.read_csv("advertising.csv")
```

```python
df1.columns
```

```python
df1['TV'] = df1['TV'].fillna(df1['TV'].mean())  #No missing values TV column, just following the procudure
```

```python
df1
```

```python
df1.isnull().sum()
```

```python
x = df1.iloc[:,:3]
```

```python
x
```

```python
y = df1['Sales']
```

```python
scaler = StandardScaler()
```

```python
arr = scaler.fit_transform(x)
```

```python
df2 = pd.DataFrame(arr)
```

```python
df2
```

```python
df2.describe()
```

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_df = pd.DataFrame()
```

```python
vif_df['vif'] = [variance_inflation_factor(arr,i) for i in range(arr.shape[1])]
vif_df['feature'] =x.columns
```

```python
vif_df
```

```python
x_train, x_test, y_train, y_test = train_test_split(arr, y,test_size=0.25,random_state=345)
```

```python
x_train
```

```python
lr = LinearRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.score(x_test,y_test)
```

```python
test = scaler.transform([[280.4,34.6,70.45]])
```

```python
lr.predict(test)
```

```python
def adj_r2(x,y):
    r2 = lr.score(x,y)
    n = x.shape[0]
    p = x.shape[1]
    adj_r2 = 1-((1-r2)*(n-1)/(n-p-1))
    return adj_r2
```

```python
adj_r2(x_test,y_test)
```

```python
lassocv = LassoCV(cv=10, max_iter=2000)
```

```python
lassocv.fit(x_train,y_train)
```

```python
lassocv.alpha_
```

```python
lasso = Lasso(alpha=lassocv.alpha_)
lasso.fit(x_train, y_train)
```

```python
lasso.score(x_test,y_test)
```

```python
ridgecv = RidgeCV(alphas = np.random.uniform(0,10,50), cv=10)
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_lr = Ridge(alpha = ridgecv.alpha_)
```

```python
ridge_lr.fit(x_train, y_train)
```

```python
ridge_lr.score(x_test,y_test)
```

```python
elasticcv = ElasticNetCV(cv = 10)
```

```python
elasticcv.fit(x_train,y_train)
```

```python
elasticcv.alpha_
```

```python
elastic_lr = ElasticNet(alpha = elasticcv.alpha_)
```

```python
elastic_lr.fit(x_train,y_train)
```

```python
elastic_lr.score(x_test,y_test)
```
